```python
import pandas as pd
```

```python
data = pd.read_excel(r"C:\Users\johan\Anaconda3\envs\minimal_ds\conda-meta\os\Data Cleaning Practice\Data - Survey Monkey Output Edited.xlsx", sheet_name= "Edited Data")
data.head(10)
```

```python
# drop unnecessary data
data_less_column = data.drop(columns = ['Start Date-','End Date-','Email Address-','First Name-','Last Name-','Custom Data 1-'])
data_less_column.head(10)
```

```python
# differentiate the id_char and the var_char
id_char = list(data_less_column)[:8]
var_char = list(data_less_column)[8:]
```

```python
data_melted = data_less_column.melt(id_vars= id_char, value_vars= var_char, var_name= 'Question + SubQuestion', value_name= 'Answer')
data_melted
```

```python
# we will merge the data_melted dataset with the question sheet in the original file
# first we will open up the question sheet in the original file
question = pd.read_excel(r"C:\Users\johan\Anaconda3\envs\minimal_ds\conda-meta\os\Data Cleaning Practice\Data - Survey Monkey Output Edited.xlsx", sheet_name= "Question")
question
```

```python
# then we could drop any column that we dont use
question.drop(columns=['Raw-Question','Raw-SubQuestion','SubQuestion','Unnamed: 5'], inplace=True)
question.head()
```

```python
# now after we clean the question data, we could merge the question data into the original data
data_merge = pd.merge(left=data_melted, right=question, how='left')
data_merge
```

```python
# now we want to know the number of respondent who answer each question
respondents = data_merge[data_merge['Answer'].notna()]
respondents = respondents.groupby('Question')['Respondent ID-'].nunique().reset_index()

# now we will rename the column of Respondent ID- into something else, because we will merge this dataset into the data_merge dataset
respondents = respondents.rename(columns={'Respondent ID-':'Respondents'})
```

```python
# and then we will merge the number of respondents with the data_merge table
data_merge_two = pd.merge(left=data_merge, right=respondents, how='left', left_on='Question', right_on='Question')
data_merge_two
```

```python
# now we want to know how many respondents that having the same answer
same_answer = data_merge
same_answer = same_answer.groupby(['Question + SubQuestion','Answer'])['Respondent ID-'].nunique().reset_index()
same_answer = same_answer.rename(columns= {'Respondent ID-':'Same Answer'})
same_answer
```

```python
# now we can merge the same_answer dataframe with the data_merge_two dataframe
data_merge_three = pd.merge(left=data_merge_two, right=same_answer, how='left', left_on=['Question + SubQuestion', 'Answer'], right_on=['Question + SubQuestion', 'Answer'])
data_merge_three
```

```python
# there are some value in the Same Answer column that is NaN, we could fill the NaN into 0  
data_merge_three['Same Answer'].fillna(value=0, inplace=True)
data_merge_three
```

```python
# now we could just clean the header name of the column
data_merge_three.rename(columns={'Identify which division you work in.-Response':'Division Primary','Identify which division you work in.-Other (please specify)':'Division Secondary','Which of the following best describes your position level?-Response':'Position','Which generation are you apart of?-Response':'Generation','Please select the gender in which you identify.-Response':'Gender','Which duration range best aligns with your tenure at your company?-Response':'Tenure','Which of the following best describes your employment type?-Response':'Employment Type'}, inplace=True)
data_merge_three
```

```python
# and last we could export this jupyter into excel
data_merge_three.to_excel(r"C:\Users\johan\Anaconda3\envs\minimal_ds\conda-meta\os\Data Cleaning Practice\Final Output.xlsx")
```
