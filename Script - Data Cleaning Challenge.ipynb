```python
import pandas as pd
```

```python
data = pd.read_csv('data_cleaning_challenge.csv')
data.head(10)
# First I will put the second row into the header of the table
# I change the header through a simple copy and paste in the csv file, so that the file could be showned as below
```

```python
# next I try to drop all the unnecessary column such as Unnamed 9 and Unnamed 10
drop_extra = data.drop(columns=['Unnamed: 9','Unnamed: 10'])
drop_extra.head(10)
```

```python
# and then I will drop all the NaN or Null values
drop_nan = drop_extra[drop_extra['Row Type'].notna()]
drop_nan.head(10)
```

```python
# I will start dividing the dataset into several part
# based on rough observation, we could see there is some characteristic in the data
# this dataset will have repeat behaviour after they encounter the row 'first name', so we will make iteration based on that condition
columns_value = []
counter = 0

for i in drop_nan['Row Type']:
    if 'first name' in i:
        counter += 1
    columns_value.append(counter)
```

```python
iter_cols = drop_nan
iter_cols['Iteration'] = columns_value
iter_cols.head(10)
```

```python
# now we are going to clean the extra rows
# we could see there is a row with Row Type, Iter Number, Power 1, Speed1, Speed 2, Electricity, Effort, Weight and, Torque
# we will select all the data that doesnt have Row Type in the Row Type column 
drop_extra_rows = iter_cols[iter_cols['Row Type'] != 'Row Type']
drop_extra_rows.head(10)
```

```python
# we will select rows with the first name, last name, date, Nan, Nan, Nan, Nan, Nan, then we are going make move the data contain with those things in new dataframe
# since we are going to clean the first name, last name, date, etc, we are going to drop the speed1, speed 2, electricity, and etc column
name_dataframe = drop_extra_rows[drop_extra_rows['Row Type'].str.contains("first name")]
name_dataframe.drop(['Speed1','Speed2','Electricity','Effort','Torque', 'Weight'], axis=1, inplace = True)

# now we will change the column name into first name, last name, and date
name_dataframe.rename({'Row Type':'First Name', 'Iter Number': 'Last Name', 'Power1':'Date'},axis=1, inplace = True)

# and then we will transform the data name from first name: person --> person
# last name: human --> human
# date: end of time --> date
name_dataframe['First Name'] = name_dataframe['First Name'].str[12:]
name_dataframe['Last Name'] = name_dataframe['Last Name'].str[11:]
name_dataframe['Date'] = name_dataframe['Date'].str[6:]
name_dataframe.head(10)
```

```python
# the remaining column will be stored in the new dataframe as well
no_name_dataframe = drop_extra_rows[~drop_extra_rows['Row Type'].str.contains("first name")]
no_name_dataframe.head(10)
```

```python
# we will merge the dataframe of the the first table and the second one on the identical/similar column such as Iteration
merge_dataframe = pd.merge(left = name_dataframe, right = no_name_dataframe, how = 'inner', on = 'Iteration')
merge_dataframe.head(10)
```
